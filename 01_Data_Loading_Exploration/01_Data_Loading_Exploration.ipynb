```python
# Import the tools we need
import pandas as pd  # pd is how we work with tabular data
import matplotlib.pyplot as plt  # matplotlib is for plots/graphs
import seaborn as sns  # seaborn makes pretty plots
```

```python
# Load the data from the CSV file
# This reads train.csv and stores it as a variable called 'df' (short for 'dataframe')
df = pd.read_csv('train.csv')
```

```python
# Print the first 5 rows so you can see what the data looks like
print("First 5 rows of the data:")
print(df.head())
```

```python
# Print the shape: how many rows and columns do we have?
print(f"\nData shape: {df.shape[0]} rows, {df.shape[1]} columns")
```

```python
# Print column names and their data types (is it a number? text? date?)
print("\nColumn info:")
print(df.info())
```

```python
# Print basic statistics (mean, median, min, max for numbers)
print("\nBasic statistics:")
print(df.describe())
```

```python
# Check for missing values (empty cells) in each column
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# Visualize missing values as a bar chart
plt.figure(figsize=(10, 4))
```

```python
df.isnull().sum().plot(kind='bar', color='red')
```

```python
plt.title('Missing Values by Column')
```

```python
plt.xlabel('Column Name')
plt.ylabel('Number of Missing Values')
```

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
# Visualize missing values as a bar chart
plt.figure(figsize=(10, 4))
df.isnull().sum().plot(kind='bar', color='red')
plt.title('Missing Values by Column')
plt.xlabel('Column Name')
plt.ylabel('Number of Missing Values')
plt.tight_layout()
plt.show()
```
